```python
import pandas as pd

# Load your dataset
df = pd.read_csv('../data/dataset.csv')

# Define a function that checks if 'None of the above' is in the value
def check_services(value):
    if 'None of the above' in value:
        return 'None of the above'
    else:
        return value

# Apply the function to the 'Services_Used' column
df['Services_Used'] = df['Services_Used'].apply(check_services)

# Save the modified data to a new CSV file
df.to_csv('../data/na.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/na.csv')

# Split the 'Services_Used' column into separate columns
services_df = df['Services_Used'].str.split(',', expand=True)

# One-hot encode the services columns
services_df = pd.get_dummies(services_df, prefix='', prefix_sep='').groupby(level=0, axis=1).sum()

# Concatenate the original DataFrame with the services DataFrame
df = pd.concat([df, services_df], axis=1)

# Drop the original 'Services_Used' column
df = df.drop('Services_Used', axis=1)

# Save the preprocessed data to a new CSV file
df.to_csv('../data/modified_na.csv', index=False)
```
